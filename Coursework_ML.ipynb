# 1. Описание данных

На первом этапе я произвела подготовку данных для анализа, которая включила: 
1. Перевод категориальных переменных в количественные путем кодирования
2. Перевод всех числовых колонок в соответствующий тип
3. Заполнение отсутствующих значений

После подготовки были рассмотрены все переменные, построены графики их распределения и посчитаны основные метрики

```python
#импортируем необходимые библиотеки:
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis
from scipy.stats import skew
```

```python
#загрузим данные в датафрейм и посмотрим его:
telecom_users_original = pd.read_csv('telecom_users.csv')
telecom_users_original.head()
```

```python
#переведем название колонок в нижний регистр
telecom_users_original.columns = map(str.lower, telecom_users_original.columns)
```

```python
#посмотрим типы данных и наличие Nan
telecom_users_original.info()
```

```python
telecom_users = telecom_users_original.copy()
```

```python
#переведем категориальные переменные (с двумя возможными значениями) в количественные путем кодирования:
telecom_users['churn'] = telecom_users['churn'].map({"No" : 0, "Yes" : 1}).astype(int)
telecom_users['paperlessbilling'] = telecom_users['paperlessbilling'].map({"No" : 0, "Yes" : 1}).astype(int)
telecom_users['partner'] = telecom_users['partner'].map({"No" : 0, "Yes" : 1}).astype(int)
telecom_users['dependents'] = telecom_users['dependents'].map({"No" : 0, "Yes" : 1}).astype(int)
telecom_users['phoneservice'] = telecom_users['phoneservice'].map({"No" : 0, "Yes" : 1}).astype(int)
telecom_users['gender'] = telecom_users['gender'].map({"Female" : 0, "Male" : 1}).astype(int)
```

```python
#для категориальных переменных с более, чем двумя возможными значениями переведем в числовую с кодировкой 0 и 1, где
# 0 - кодирует одновременно отсутствие сервиса и отсутствие данной дополнительной услуги, а 1 - наличие сервиса и услуги
telecom_users['multiplelines'] = telecom_users['multiplelines'].map({"No" : 0, "Yes" : 1, "No phone service" : 0}).astype(int)
telecom_users['onlinesecurity'] = telecom_users['onlinesecurity'].map({"No" : 0, "Yes" : 1, "No internet service" : 0}).astype(int)
telecom_users['onlinebackup'] = telecom_users['onlinebackup'].map({"No" : 0, "Yes" : 1, "No internet service" : 0}).astype(int)
telecom_users['deviceprotection'] = telecom_users['deviceprotection'].map({"No" : 0, "Yes" : 1, "No internet service" : 0}).astype(int)
telecom_users['techsupport'] = telecom_users['techsupport'].map({"No" : 0, "Yes" : 1, "No internet service" : 0}).astype(int)
telecom_users['streamingtv'] = telecom_users['streamingtv'].map({"No" : 0, "Yes" : 1, "No internet service" : 0}).astype(int)
telecom_users['streamingmovies'] = telecom_users['streamingmovies'].map({"No" : 0, "Yes" : 1, "No internet service" : 0}).astype(int)
```

```python
#для категориальной переменной "тип контракта" с понятным порядком кодируем значения по порядку
telecom_users['contract'] = telecom_users['contract'].map({"Month-to-month" : 0, "One year" : 1, "Two year" : 2}).astype(int)
```

```python
#для категориальной переменной "услуга интернета" кодируем 0 отсутствие услуги, 1 - наличие услуги 
#(одинаково для типа соединения)
telecom_users['internetservice'] = telecom_users['internetservice'].map({"No" : 0, "DSL" : 1, "Fiber optic" : 1}).astype(int)
```

```python
#переводим строковую колонку 'totalcharges' в числа с плавающей точкой и заполняем Nan нулями (поскольку это новые клиенты)
telecom_users['totalcharges'] = pd.to_numeric(telecom_users['totalcharges'], errors='coerce')
telecom_users['totalcharges'] = telecom_users['totalcharges'].fillna(0)
```

Перед тем, как приводить данные к одному масштабу, посмотрим на характер их распределения. Большинство переменных в нашем датасете имеют дискретный характер, сначала построим графики для этих переменных. 

```python
#создадим два списка - с названиями колонок с биноминальными переменными и с подписями оси x для каждой из них
binom_columns = ['gender', 'seniorcitizen', 'partner', 'dependents', 'phoneservice', 'multiplelines', 'internetservice', 
                'onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies', 
                'paperlessbilling']
x_labels = [['Male', 'Female'], ['Non-seniorcitizen', 'Seniorcitizen'], ['No_partner', 'Is_partner'], 
        ['No_dependents', 'Is_dependents'], ['No_phoneservice', 'Is_phoneservice'], 
        ['No_multiplelines', 'Is_multiplelines'], ['No_internetservice', 'Is_internetservice'], 
        ['No_onlinesecurity', 'Is_onlinesecurity'], ['No_onlinebackup', 'Is_onlinebackup'], 
        ['No_deviceprotection', 'Is_deviceprotection'], ['No_techsupport', 'Is_techsupport'], 
        ['No_streamingtv', 'Is_streamingtv'], ['No_streamingmovies', 'Is_streamingmovies'], 
        ['No_paperlessbilling', 'Is_paperlessbilling']]
```

```python
# нарисуем диаграммы для всех переменных в цикле
for column, x_label in zip(binom_columns, x_labels):
    fig, ax = plt.subplots()
    ax.bar(x_label, telecom_users.groupby(column)[column].agg('count').values, color='b')
    ax.bar(x_label, telecom_users.groupby(
        [column, 'churn'])['churn'].agg('count').values.reshape(2, 2)[:, 1], color='c')

    #дополнительно создадим подписи данных
    rects = ax.patches
    labels = np.around(telecom_users.groupby(
        [column, 'churn'])['churn'].agg('count').values.reshape(2, 2)[:, 1] * 100 / telecom_users.groupby(
        column)[column].agg('count').values, 1)
    for rect, label in zip(rects, labels):
        height = rect.get_height()*0.005
        ax.text(rect.get_x() + 0.4, height, f'{str(label)}%', ha='center', va='bottom', color='w')
    
    #вставим заголовок
    ax.set_title(f'Переменная {column}')
    
    ax.legend(['Non-Churn', 'Churn'])
```

Описание переменных:
1. Пол - выборка сбалансирована по возрасту клиентов, разница по доле оттока среди мужчин и женщин менее 1 п.п.
2. Пенсионеры - основная часть клиентов выборки не являются пенсионерами. Доля оттока выше среди пенсионеров
3. Наличие партнера - количество клиентов с и без партнера сопоставимо в выборке. Доля оттока выше среди клиентов без партнера.
4. Наличие иждивенцев - в выборке преобладают клиенты без иждивенцев, доля оттока среди них больше.
5. Телефонный сервис есть у основного числа клиентов. Доля оттока среди тех, у кого есть сервис и у кого нет отличается на 1,3 п.п.
6. Многоканального сервиса нет у большинства клиентов, разница в оттоке незначительна (3,3 п.п.)
7. У большинства клиентов есть интернет сервис, доля оттока по таким клиентам выше (на 24 п.п.)
8. По остальным интернет-услугам распределение примерно одинаковое - большая часть клиентов услугами не пользуется. Что касается отличий по оттоку, то значимая разница наблюдается по следующим услугам: доля оттока выше на 17 п.п. среди тех, кто не подключает услугу "onlinesecurity"; на 15.8 п.п. среди тех, кто не подключает услугу "techsupport"
9. Услуга "paperlessbilling" есть у большинства клиентов, доля оттока среди подключивших эту услугу выше, чем по тем, кто не подключил на 17 п.п.

Таргет представляет собой биноминальную переменную, поэтому мы имеем задачу **двухклассовой классификации**. Мы видим, что выборка несбалансирована по целевой переменной, в следствии чего при оценке качества модели **целесообразно использовать метрику roc_auc**.

Теперь перейдем к описанию непрерывных переменных - monthlycharges, totalcharges и tenure

```python
#напишем функцию для построения гистограммы и распечатки базовых метрик для непрерывных переменных:
def my_basic_research(df, column, bins):
    print(f'Переменная {column}\n')
    print(f'Базовые метрики переменной {column} \n{df[column].describe()}\n')
    print(f'Медиана {round(df[column].median(), 4)}')
    print(f'Эксцесс {round(kurtosis(df[column]), 4)}')
    print(f'Ассиметрия {round(skew(df[column]), 4)}')
    fig, ax = plt.subplots()
    ax.hist(df[column].values, bins=bins, align='mid')
    ax.set_title(f'Распределение переменной {column}')
```

```python
#смотрим данные для месячного размера оплаты услуг
my_basic_research(telecom_users, telecom_users.columns[19], 10)
```

Пользователей условно можно разделить на два кластера - с низкой стоимостью услуг и с высокой стоимостью. Среднее значение в таком распределении не отражает его характер.

```python
#смотрим данные для полного размера оплаты услуг
my_basic_research(telecom_users, telecom_users.columns[20], 5)
```

Распределение переменной полной оплаты услуг за все время пользования имеет длинный правый хвост, большинство пользователей сосредоточено в диапазоне с низкими значениями рассматриваемого показателя 

```python
#смотрим данные для показателя времени пользования услугами компании
my_basic_research(telecom_users, telecom_users.columns[6], 5)
```

Здесь также наблюдается два кластера - новые пользователи (менее года пользуются услугами) и старые пользователи (пользуются услугами более 5 лет)

```python
#пишем функцию для отрисовки ящика с усами
def boxplot_research(df, column):
    fig, ax = plt.subplots()
    ax.boxplot(df[column].values, meanline=True)
    ax.set_title(f'Распределение переменной {column}')
```

```python
#применяем функцию для каждого столбца с непрерывной переменной
fig, ax = plt.subplots()
ax.boxplot([telecom_users['monthlycharges'], telecom_users['tenure']], meanline=True)
ax.set_title(
    f"Распределение переменных '{telecom_users.columns[19]}' и {telecom_users.columns[6]}'"
)
```

```python
boxplot_research(df=telecom_users, column=telecom_users.columns[20])
```

# 2. Исследование зависимостей и формулирование гипотез

На втором этапе я изучила зависимости между переменными, сформулировала гипотезы подготовила данные для передачи их в модель (создала фиктивные переменные, привела их к одному масштабу, понизила размерность коррелирующих между собой переменных).

# 3. Построение моделей для прогнозирования оттока

В рамках третьей части работы на основании выводов 1-го и 2-го разделов было принято решение **разделить выборку на три части** - обучающую, валидационную и тестовую. На обучающей и валидационной выборке выберем лучший набор признаков, на тестовой - сравним алгоритмы на этом наборе. Будем использовать **GridSearchCV** для подбора гиперпараметров моделей. В качестве базовых алгоритмов используем **логистическую регрессию** (далее - LR) и **метод опорных векторов** (далее - SVC), поскольку между признаками и таргетом наблюдается линейная зависимость. Также используем алгоритмы решающих деревьев: **Gradient Boosting Classifier** (далее - GBC) и **Random Forest Classifier** (далее - RFC) для сравнения. В качестве метрики будем везде использовать **roc_auc_score**, поскольку классы у нас не сбалансированы. В случае наличия сомнений при выборе модели можно ориентироваться дополнительно на метрику recall, поскольку в данной задаче более важным является минимизация ложноотрицательных срабатываний (алгоритм предсказывает, что оттока нет (класс 0), а на самом деле, он есть (класс 1).

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score, recall_score
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold)
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from xgboost import XGBClassifier

# задаем параметры базовых алгоритмов
gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)} # GradientBoostingClassifier

rfc_params = {'n_estimators': range(10, 100, 10), # RandomForestClassifier
              'min_samples_leaf': range(1, 5)}

svc_params = {'kernel': ['linear', 'rbf'], # SVC
              'C': np.arange(0.1, 1, 0.2)}

lr_params = {'C': np.arange(0.5, 1, 0.1)} # Logistic regression

skf = StratifiedKFold(n_splits=5, random_state=17)
```

```python
#прописываем модели для обучения
gbc = GradientBoostingClassifier(random_state=17)
rfc = RandomForestClassifier(class_weight={0:.35, 1:.65}, random_state=17)
svc = SVC(class_weight={0:.35, 1:.65}, random_state=17)
lr = LogisticRegression(class_weight={0:.35, 1:.65}, random_state=17)


gbc_grid = GridSearchCV(gbc, gbc_params, scoring='roc_auc', cv=skf)
rfc_grid = GridSearchCV(rfc, rfc_params, scoring='roc_auc', cv=skf)
svc_grid = GridSearchCV(svc, svc_params, scoring='roc_auc', cv=skf)
lr_grid = GridSearchCV(lr, lr_params, scoring='roc_auc', cv=skf)

#в цикле делим данные для каждого набора данных, обучаем модели по каждому набору на тренировочной выборке и
#делаем предсказания на валидационной выборке, рассчитываем roc_auc_score, сохраняем в словарь.
metrics = []

for set_of_features in sets_of_features:
    X = set_of_features.drop(['churn'], axis = 1).values
    y = set_of_features['churn'].values
    data_train, data_test, target_train, target_test = train_test_split(X, y, random_state=17, test_size=.3)
    x_train, x_valid, y_train, y_valid = train_test_split(data_train, target_train, random_state=17, test_size=.3)
    
    gbc_grid_model = gbc_grid.fit(x_train, y_train)
    rfc_grid_model = rfc_grid.fit(x_train, y_train)
    svc_grid_model = svc_grid.fit(x_train, y_train)
    lr_grid_model = lr_grid.fit(x_train, y_train)
    
    gbc_grid_predict = gbc_grid.predict(x_valid)
    rfc_grid_predict = rfc_grid.predict(x_valid)
    svc_grid_predict = svc_grid.predict(x_valid)
    lr_grid_predict = lr_grid.predict(x_valid)
    
    metrics.append([roc_auc_score(y_valid, gbc_grid_predict), roc_auc_score(y_valid, rfc_grid_predict), roc_auc_score(y_valid, svc_grid_predict), roc_auc_score(y_valid, lr_grid_predict)])
```

```python
#добавим рассчитанные метрики по каждому набору признаков в датафрейм
model_metrics = pd.DataFrame(metrics,
    index=['SVD_1', 'SVD_2', 'SVD_3', 'noSVD_1', 'noSVD_2', 'noSVD_3'], 
    columns=['GBC', 'RFC', 'SVC', 'LR']
)
model_metrics
```

Наибольшие значения метрики roc_auc у набор данных noSVD_3 - 0.761632

```python
#теперь посчитаем результаты моделей на тестовом датасете, снова разделим выбранный набор признаков (noSVD_3) на три части
X = telecom_users_nosvd_3.drop(['churn'], axis = 1).values
y = telecom_users_nosvd_3['churn'].values

data_train, data_test, target_train, target_test = train_test_split(X, y, random_state=17, test_size=.3)
x_train, x_valid, y_train, y_valid = train_test_split(data_train, target_train, random_state=17, test_size=.3)

#передадим в модели наилучшие параметры:  
gbc = GradientBoostingClassifier(**gbc_grid.best_params_, random_state=17)
rfc = RandomForestClassifier(**rfc_grid.best_params_, class_weight={0:.35, 1:.65}, random_state=17)
svc = SVC(**svc_grid.best_params_, class_weight={0:.35, 1:.65}, random_state=17)
lr = LogisticRegression(**lr_grid.best_params_, class_weight={0:.35, 1:.65}, random_state=17)

#обучим модели с выбранным набором
gbc_grid_model = gbc.fit(x_train, y_train)
rfc_grid_model = rfc.fit(x_train, y_train)
svc_grid_model = svc.fit(x_train, y_train)
lr_grid_model = lr.fit(x_train, y_train)

#сделаем предсказание на тестовом наборе
gbc_grid_predict = gbc.predict(data_test)
rfc_grid_predict = rfc.predict(data_test)
svc_grid_predict = svc.predict(data_test)
lr_grid_predict = lr.predict(data_test)
    
metrics_test = [roc_auc_score(target_test, gbc_grid_predict), roc_auc_score(target_test, rfc_grid_predict), roc_auc_score(target_test, svc_grid_predict), roc_auc_score(target_test, lr_grid_predict)]
```
